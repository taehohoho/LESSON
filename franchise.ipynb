```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# 가맹사업정보시스템 URL.
target_url = "https://franchise.ftc.go.kr/main/index.do"
```

```python
# 크롬 드라이버 로딩.
driver = webdriver.Chrome()
driver.get(target_url)
```

```python
# 브랜드별 비교 탭 누르기.
driver.find_element_by_css_selector(".main_tab01 .tab03 > a").click()
```

```python
# 치킨 선택.
driver.find_element_by_css_selector("#selIndus3 > option[value='H1']").click()
```

```python
# 가맹점 현황 정보 선택.
driver.find_element_by_css_selector("select[name='selListType'] > option[value='listBrand02']").click()
```

```python
# 검색 버튼 클릭.
driver.find_element_by_css_selector("#frmSearch3").submit()
```

```python
# 내림차순 버튼 클릭.
driver.find_element_by_css_selector(".tdown").click()
```

```python
# 전체 페이지 소스 얻어오기.
source = driver.page_source
driver.close()
source[:100]
```

```python
# 페이지 소스를 BeautifulSoup에 연결.
soup = BeautifulSoup(source, "html.parser")
```

```python
# 행 리스트 가져오기.
rows = soup.select(".table tbody tr")[1:]
rows[0]
```

```python
# 이름 가져오기.
names = [row.select_one("td:first-child").string for row in rows]
names
```

```python
# 가맹점수 가져오기.
number_of_stores = [int(row.select_one("td:nth-child(3)").string.replace(",", "")) for row in rows]
number_of_stores
```

```python
# 가맹점 평균매출액 가져오기.
average_sales = [int(row.select_one("td:nth-child(8)").string.strip().split(" ")[0].replace(",", "")) for row in rows]
average_sales
```

```python
# 가져온 데이터를 바탕으로 데이터프레임 만들기.
data = pd.DataFrame({
    "브랜드명": names,
    "가맹점 수": number_of_stores,
    "가맹점 평균매출액": average_sales
})
data
```

```python
# 중복된 항목 제거.
data.drop_duplicates(subset = "브랜드명", keep = False, inplace = True)
data.reset_index(drop = True, inplace = True)
data
```

```python
# 가맹점 수가 0인 행 제거.
data.drop(data.loc[data["가맹점 수"] == 0].index, inplace = True)
data
```

```python
# 가맹점 평균매출액이 0인 행 제거.
data.drop(data.loc[data["가맹점 평균매출액"] == 0].index, inplace = True)
data.reset_index(drop = True, inplace = True)
data
```

```python
# 목표 브랜드명으로 필터링. (맘스터치, 푸라닭 제외)
target_brands = ["비비큐(BBQ)", "굽네치킨", "비에이치씨(BHC)", "교촌치킨", "노랑통닭", "60계", "바른치킨", "자담치킨", "네네치킨", "호식이두마리치킨", "지코바양념치킨", "처갓집양념치킨", "깐부치킨", "페리카나", "또래오래", "아웃닭", "땅땅치킨", "또봉이통닭"]
data = data.loc[data["브랜드명"].isin(target_brands)]
data.reset_index(drop = True, inplace = True)
data
```

```python
# 맘스터치와 푸라닭 데이터 삽입. (출처: 구글링)
data = data.append({
    "브랜드명": "맘스터치",
    "가맹점 수": 1348,
    "가맹점 평균매출액": None
}, ignore_index = True)
data = data.append({
    "브랜드명": "푸라닭",
    "가맹점 수": 241,
    "가맹점 평균매출액": None
}, ignore_index = True)
data = data.sort_values(by = "가맹점 수", ascending = False)
data.reset_index(drop = True, inplace = True)
data
```

```python
# 엑셀 파일로 내보내기.
data.to_excel("./franchise.xlsx")
```
