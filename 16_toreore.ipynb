```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

```python
# webdriver로 크롬 브라우저 실행하기
browser = webdriver.Chrome('C:/Myexam/chromedriver/chromedriver.exe')
```

```python
#  반복문으로 유튜브 랭킹 화면의 여러 페이지를 크롤링하기
results = []
for page in range(1,16):
    url = f"https://pelicana.co.kr/store/stroe_search.html?page={page}&branch_name=&gu=&si=%EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C"
    browser.get(url)
    time.sleep(2)
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')
    lists = soup.select('section > table > tbody > tr') 
    for store in lists:
        name = store.select('td')[0].text
        address = store.select('td')[1].text
        data = [name, address]
        results.append(data)
results
```

```python
# 데이터 칼럼명을 설정하고 엑셀 파일로 저장하기
df =  pd.DataFrame(results)
df.columns = ['name', 'address']
df.to_excel('./17_toreore.xlsx', index = False )
```
